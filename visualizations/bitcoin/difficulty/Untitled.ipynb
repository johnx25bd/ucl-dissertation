```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('./difficulty.csv')
```

```python
df['date'] = pd.to_datetime(df['2009-01-03 00:00:00'])
```

```python
df['bits'] = np.log2(df['1.0'])
```

```python
df
```

```python

fig, ax = plt.subplots(figsize=(10,4))

ax.plot(df['date'], df['1.0'])

# Rewrite the y labels
y_labels = ax.get_yticks()
ax.yaxis.get_major_formatter().set_scientific(False)

ax.title.set_text("Bitcoin network block difficulty")
ax.set_xlabel('Date')
ax.set_ylabel(r'Block difficulty, $2^{n  bits}$')

# plt.savefig('./bitcoin-block-difficulty.svg')
# plt.savefig('./bitcoin-block-difficulty.png')
```
