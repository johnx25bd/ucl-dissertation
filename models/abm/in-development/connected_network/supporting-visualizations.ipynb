```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

plt.style.use('seaborn-muted')
```

## Visualizing Ethereum block confirmation times

Data from https://etherscan.io/chart/blocktime.

```python
df_blocktimes = pd.read_csv('../../../../visualizations/ethereum/block-confirmations/export-BlockTime.csv')
df_blocktimes['date'] = pd.to_datetime(df_blocktimes['Date(UTC)'])
```

```python
date_format = DateFormatter("%b %Y") 

# plot the data
fig, ax = plt.subplots()

ax.plot(df_blocktimes['date'], 
       df_blocktimes['Value'])

ax.axhline( df_blocktimes['Value'].mean(), color='red', linestyle='--')
ax.set(xlabel="Date", ylabel="Mean daily block confirmation time (seconds)")
ax.set(title="Ethereum mainnet block confirmation times (seconds) by date")

ax.text(df_blocktimes.loc[1325, 'date'], 16.1, 
        'Mean = ' + str(round(df_blocktimes['Value'].mean(), 3)),
       color='red')

# tell matplotlib to use the format specified above
ax.xaxis.set_major_formatter(date_format); 
fig.set_size_inches(10, 4)

# fig.savefig('../../../../visualizations/ethereum/block-confirmations/ethereum-block-times.svg')
```
